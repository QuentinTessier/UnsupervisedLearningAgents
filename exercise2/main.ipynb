```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

data = np.load('data.npy')
labels = np.load('labels.npy')
```

Initialisation des différents techniques de réduction dimentionelle

```python
pca_2d = PCA(n_components=2)
pca_3d = PCA(n_components=3)

tsne_2d = TSNE(n_components=2)
tsne_3d = TSNE(n_components=3)

pca_data_2d = pca_2d.fit_transform(data)
pca_data_3d = pca_3d.fit_transform(data)
tsne_data_2d = tsne_2d.fit_transform(data)
tsne_data_3d = tsne_3d.fit_transform(data)
```

Plotting des resultats des différents réduction

```python
fig = plt.figure(figsize=(6, 6))
f2 = fig.add_subplot()
f2.scatter(pca_data_2d[:, 0], pca_data_2d[:, 1], c=labels)
plt.show()

fig = plt.figure(figsize=(6, 6))
f1 = fig.add_subplot(projection='3d')
f1.scatter(pca_data_3d[:, 0], pca_data_3d[:, 1], pca_data_3d[:, 2], c=labels)
plt.show()

fig = plt.figure(figsize=(6, 6))
f2 = fig.add_subplot()
f2.scatter(tsne_data_2d[:, 0], tsne_data_2d[:, 1], c=labels)
plt.show()

fig = plt.figure(figsize=(6, 6))
f1 = fig.add_subplot(projection='3d')
f1.scatter(tsne_data_3d[:, 0], tsne_data_3d[:, 1], tsne_data_3d[:, 2], c=labels)
plt.show()
```

La technique PCA semble être la plus performante dans notre cas. En effet, les 2 blobs de donnée se superpose moins que lors de l'utilisation de TSNE.

La PCA en 2 dimension semble permettre de prédire le label.
